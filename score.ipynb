```python
rouge_1 = []
rouge_2 = []
rouge_L = []
scorer = rouge_scorer.RougeScorer(['rouge1','rouge2', 'rougeL'], use_stemmer=True)
for i in range(valid.shape[0]):
    scores = scorer.score(valid['summary'][i],
                      valid['summary'][i])
    print(scores['rouge1'].fmeasure)
    rouge_1.append(scores['rouge1'].fmeasure)
    rouge_2.append(scores['rouge2'].fmeasure)
    rouge_L.append(scores['rougeL'].fmeasure)
```

```python
rouge_1_score = sum(rouge_1)/valid.shape[0]
rouge_2_score = sum(rouge_2)/valid.shape[0]
rouge_L_score = sum(rouge_L)/valid.shape[0]
```

```python
print(rouge_1_score, rouge_2_score, rouge_L_score)
```
